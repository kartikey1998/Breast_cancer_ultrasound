```python
from config_module import CFG as config
import model_module
import data_module
import utils_module

import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Prints the size of the three data splits
print(f"size of Test split: {len(data_module.df_test)}")
print(f"size of Validation split: {len(data_module.df_valid)}")
print(f"size of train split: {len(data_module.df_main)}")
```

```python
# Initialize segmentation model
unet_segment = model_module.UNet()
lossFunc_segment = nn.BCEWithLogitsLoss()
opt_segment = optim.Adam(unet_segment.parameters(), lr=config.LEARNING_RATE, weight_decay=config.WEIGHT_DECAY)


# Initialize classification model
resnet_classifier = model_module.Res_Net()
lossFunc_classify = nn.BCEWithLogitsLoss()
opt_classify = optim.Adam(resnet_classifier.parameters(), lr=config.LEARNING_RATE, weight_decay=config.WEIGHT_DECAY)
```

```python
history_classfy = model_module.train_classifier(resnet_classifier, lossFunc_classify, opt_classify, config.NUM_EPOCHS, 
                           data_module.train_dataloader, data_module.valid_dataloader)
```

```python
utils_module.plot_training_loss(history_classfy)
```

```python
utils_module.plot_training_Accuracy(history_classfy)
```

```python
history_seg = model_module.train_segmentation_model(unet_segment, lossFunc_segment, opt_segment,config.NUM_EPOCHS,
                                                data_module.train_dataloader, data_module.valid_dataloader)
```

```python
utils_module.plot_training_loss(history_seg)
```
